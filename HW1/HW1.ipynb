## Task 1. Compiling Ebola Data

The `DATA_FOLDER/ebola` folder contains summarized reports of Ebola cases from three countries (Guinea, Liberia and Sierra Leone) during the recent outbreak of the disease in West Africa. For each country, there are daily reports that contain various information about the outbreak in several cities in each country.

Use pandas to import these data files into a single `Dataframe`.
Using this `DataFrame`, calculate for *each country*, the *daily average per month* of *new cases* and *deaths*.
Make sure you handle all the different expressions for *new cases* and *deaths* that are used in the reports.


```python
import os
guinea_list=[]
pathname = os.path.join(DATA_FOLDER+"\\ebola", 'guinea_data')
# Concat all the DF from each csv from each country
for csv_file in os.listdir(pathname):
    abs_path=os.path.join(pathname,csv_file)
    df=pd.read_csv(abs_path,index_col=0,parse_dates=[0])
    guinea_list.append(df)
df=pd.concat(guinea_list)
df['Month']=df.index.month
df
```

```python
df_reduced=df[['Month','Description','Totals']]
interesting_metrics_description=['New cases','New deaths']

# Change "new deaths registered today" and "New deaths registered" of first day to "new deaths". same for cases
new_description=df_reduced['Description']
new_description[new_description=='New deaths registered today']='New deaths'
new_description[new_description=='New deaths registered']='New deaths'
new_description[new_description=='Total new cases registered so far']='New cases'
df_reduced['Description']=new_description

# Getting number of days in month
df_reduced['Date']=df_reduced.index
# Sort by description to get only one metric
df_reduced=df_reduced.set_index(['Description'])
# Only the cases
months=df_reduced.loc[interesting_metrics_description[0]]
# Keeping only the months
months=months['Month']
# Counting the occurences of each month (number of csv files we have per month)
months=months.value_counts()
# Creating a DF with this
months=pd.DataFrame(months)
# Change the name of Month to days per month since it's the occurences
months=months.rename(columns={'Month':'Days_per_month'})
months['Month']=months.index

# Getting relevant descriptions
df_reduced_only_cool_metrics=df_reduced.loc[interesting_metrics_description]
# Sort by description/month/date
df_reduced_only_cool_metrics=df_reduced_only_cool_metrics.reset_index().set_index(['Description','Month','Date'])
# Converting to from object to int all the integers in Totals
df_reduced_only_cool_metrics['Totals']=pd.to_numeric(df_reduced_only_cool_metrics['Totals'])
# Reduce to Description/month by summning
df_summed=df_reduced_only_cool_metrics.sum(level=['Description','Month'])
# Reset the index and merge with DF months by column Month
df_summed=df_summed.reset_index()
df_merged=pd.merge(df_summed,months,on='Month',how='left')
df_merged=df_merged.set_index(['Description','Month'])
# Divide Totals by days per month
df_merged['Totals']=df_merged['Totals']/df_merged['Days_per_month']
df_divided=df_merged.drop('Days_per_month',axis=1)
df_guinea=df_divided
df_guinea
# Check number of days
```

For Guinea, we have two different decriptions for new deaths : "New deaths registered today" and "New deaths registered", that we rename "New deaths".

For cases, there is only the description "Total new cases registered so far" that we rename "New cases".

First we calculate the number of csv files per month, so that we can divide the total number of cases/deaths per month and get the daily average. 

Then we aggregate our data per month and calculate our daily average per month.

```python
import os
liberia_list=[]
pathname = os.path.join(DATA_FOLDER+"\\ebola", 'liberia_data')
# Concat all the DF from each csv from each country
for csv_file in os.listdir(pathname):
    abs_path=os.path.join(pathname,csv_file)
    df=pd.read_csv(abs_path,index_col=0,parse_dates=[0])
    liberia_list.append(df)
df=pd.concat(liberia_list)
df['Month']=df.index.month
df
```

```python
# Keeping only the variables of interest
df_reduced=df[['Month','Variable','National']]
# Convert National column to numeric
df_reduced['National']=pd.to_numeric(df_reduced['National'])
# Are we sure ???
# df_reduced=df_reduced.dropna()

# Newly reported deaths
# New Case/s (Suspected)
# New Case/s (Probable)
# New case/s (confirmed)

# Recovering the date from the indices
df_reduced['Date']=df_reduced.index
# Sorting by decription
df_reduced_descr=df_reduced.set_index('Variable')

# Recovering the months the same way as before
months=df_reduced_descr.loc['Newly reported deaths']
months=months['Month']
months=months.value_counts()
months=pd.DataFrame(months)
months=months.rename(columns={'Month':'Days_per_month'})
months['Month']=months.index

# Reovering the new cases and summing them together (suspected probable and confirmed)
df_reduced_selected_cases=df_reduced_descr.loc[['New Case/s (Suspected)','New Case/s (Probable)','New Case/s (confirmed)']]
df_reduced_selected_cases=df_reduced_selected_cases.reset_index().set_index(['Month','Variable','Date'])
# Summing keeping only the month
df_summed_cases=df_reduced_selected_cases.sum(level=[0])
# Create the Variable new cases
df_summed_cases['Variable']=['New Cases']*df_summed_cases.shape[0]
df_summed_cases=df_summed_cases.reset_index()

# From a certain point in december, we don't have the newly reported deaths,
# So we need to recover them from total deaths in all cases.
# To do that we do the inverse of a cumulative sum by substracting the previous element
df_filler=df_reduced_descr.loc['Total death/s in confirmed, \r\n probable, suspected cases'] 

df_filler=df_filler.reset_index()
# The list we will substract
tmp=list(df_filler['National'])
# So that i is subtracted by elem in i-1
tmp[1:]=tmp[:-1]
# Adding the column that we will substract by
df_filler=pd.concat([df_filler,pd.DataFrame(list(tmp),columns=['to_remove'])],axis=1)
#substracting
df_filler['National']=df_filler['National']-df_filler['to_remove']
df_filler=df_filler.drop('to_remove',axis=1)
df_filler=df_filler.set_index('Date')
# PROBLEM: from the 31-oct to the 1-nov, the total deaths decrease by 1518. 

# New deaths
df_deaths=df_reduced_descr.loc['Newly reported deaths']
df_deaths=df_deaths.reset_index().set_index('Date')
# Filling the missing values with the inversed cumulative deaths
df_deaths['National']=df_deaths['National'].fillna(df_filler['National'])
df_deaths=df_deaths.reset_index().set_index(['Month','Variable','Date'])
# Reduce to variable/month by summing over the dates of the same months
df_summed_death=df_deaths.sum(level=[0,1])
df_summed_death=df_summed_death.reset_index()
# Creating the new dath variable
df_summed_death['Variable']=['New deaths']*df_summed_death.shape[0]
# Concat with the cases
df_summed=pd.concat([df_summed_cases,df_summed_death])
# Merge with the days per months
df_merged=pd.merge(df_summed,months,on='Month',how='left')
# Rename the col Variable by Description to be consistent with guinea
df_merged=df_merged.rename(columns={'Variable':'Description'})
df_merged=df_merged.set_index(['Description','Month'])
# Divide col National by days per month for daily average and rename it to Total col to be consistent with guinea
df_merged['Totals']=df_merged['National']/df_merged['Days_per_month']
df_divided=df_merged.drop(['Days_per_month','National'],axis=1)
df_divided=df_divided.dropna()
df_liberia=df_divided
df_liberia
```

For Liberia, we have two different variables for new deaths. We use "Newly reported deaths"  when it's available. Otherwise we use the "Total death/s in confirmed, probable, suspected cases" that we "decumulate" as it's a cumulative sum of the deaths. The final combined result becomes "New deaths".

For cases, there are three different variables that we sum together: "New Case/s (Suspected)", "New Case/s (Probable)" and "New case/s (confirmed)" become "New cases"

We also calculate the number of csv files per month, so that we can divide the total number of cases/deaths per month and get the daily average. 

Then we aggregate our data per month and calculate our daily average per month.

**In December, we get an large number of new cases. This is because in the last csv files, we have an very large amount of new cases. This might be an error and they might have confused new cases with total cases.**

```python
import os
sl_list=[]
pathname = os.path.join(DATA_FOLDER+"\\ebola", 'sl_data')
#Concat all the csv files of sl into one df
for csv_file in os.listdir(pathname):
    abs_path=os.path.join(pathname,csv_file)
    df=pd.read_csv(abs_path,index_col=0,parse_dates=[0])
    sl_list.append(df)
df=pd.concat(sl_list)
df['Month']=df.index.month
df
```

```python
# Keeping only relevant columns
df_reduced=df[['Month','variable','National']]
# Converting to numeric
df_reduced['National']=pd.to_numeric(df_reduced['National'],errors='coerce')

df_reduced['Date']=df_reduced.index
df_reduced_descr=df_reduced.set_index('variable')
# Recovering the number of days per months as before
months=df_reduced_descr.loc['death_suspected']
months=months['Month']
months=months.value_counts()
months=pd.DataFrame(months)
months=months.rename(columns={'Month':'Days_per_month'})
months['Month']=months.index

# do the inverse of the cumulative sum for death_suspected, to get new deaths per day. Same idea as for liberia
shifter1=df_reduced_descr.loc['death_suspected'] 
shifter1=shifter1.reset_index()
tmp=list(shifter1['National'])
tmp[1:]=tmp[:-1]
shifter1=pd.concat([shifter1,pd.DataFrame(list(tmp),columns=['to_remove'])],axis=1)
shifter1['National']=shifter1['National']-shifter1['to_remove']
shifter1=shifter1.drop('to_remove',axis=1)

# Same for death probable
shifter2=df_reduced_descr.loc['death_probable']  
shifter2=shifter2.reset_index()
tmp=list(shifter2['National'])
tmp[1:]=tmp[:-1]
shifter2=pd.concat([shifter2,pd.DataFrame(list(tmp),columns=['to_remove'])],axis=1)
shifter2['National']=shifter2['National']-shifter2['to_remove']
shifter2=shifter2.drop('to_remove',axis=1)

# same for death confirmed
shifter3=df_reduced_descr.loc['death_confirmed']  
shifter3=shifter3.reset_index()
tmp=list(shifter3['National'])
tmp[1:]=tmp[:-1]
shifter3=pd.concat([shifter3,pd.DataFrame(list(tmp),columns=['to_remove'])],axis=1)
shifter3['National']=shifter3['National']-shifter3['to_remove']
shifter3=shifter3.drop('to_remove',axis=1)

# regroup all the deaths
df_shifted=pd.concat([shifter1,shifter2,shifter3])

# Sum all the deaths
df_shifted=df_shifted.set_index('Month')
df_deaths=df_shifted.sum(level='Month')
df_deaths['Description']=['New deaths']*df_deaths.shape[0]
df_deaths=df_deaths.reset_index()

# Sum all the cases
df_cases=df_reduced_descr.loc[['new_suspected','new_probable','new_confirmed']]
df_cases=df_cases.reset_index().set_index('Month')
df_cases=df_cases.sum(level='Month')
df_cases['Description']=['New cases']*df_cases.shape[0]
df_cases=df_cases.reset_index()

# sum over all the days of a months
df_selected=pd.concat([df_cases,df_deaths])
df_selected=df_selected.set_index(['Description','Month'])
df_summed=df_selected.sum(level=[0,1])
df_summed=df_summed.reset_index()

# Divide by the number of days
df_merged=pd.merge(df_summed,months,on='Month',how='left')
df_merged=df_merged.set_index(['Description','Month'])
df_merged['Totals']=df_merged['National']/df_merged['Days_per_month']
df_divided=df_merged.drop(['Days_per_month','National'],axis=1)
df_divided=df_divided.dropna()
df_sl=df_divided
df_sl
```

For Sierra Leone, we have three different variables for new deaths. "death_suspected", "death_probable" and "death_confirmed" that we "decumulate" as they are a cumulative sum of the deaths. The final combined result becomes "New deaths".

For cases, there are three different variables that we sum together: 'new_suspected','new_probable' and 'new_confirmed' become "New cases".

We also calculate the number of csv files per month, so that we can divide the total number of cases/deaths per month and get the daily average. 

Then we aggregate our data per month and calculate our daily average per month.

## Task 2. RNA Sequences

In the `DATA_FOLDER/microbiome` subdirectory, there are 9 spreadsheets of microbiome data that was acquired from high-throughput RNA sequencing procedures, along with a 10<sup>th</sup> file that describes the content of each. 

Use pandas to import the first 9 spreadsheets into a single `DataFrame`.
Then, add the metadata information from the 10<sup>th</sup> spreadsheet as columns in the combined `DataFrame`.
Make sure that the final `DataFrame` has a unique index and all the `NaN` values have been replaced by the tag `unknown`.

### Description 

    - We will first load all excel files as dataFrame, next to one another in order to have for a bacteria, all the results for different conditions on the same row. The column names will first be the barcode associated and the dataFrames will be joined by an outer join to have all possible bacteria as rows. We replace the 'NaN' values by 'unknown'. 
    - Then we will load the metadata as a dataFrame of columns barcodes and bacteria, in order to concatenate it with the data. 
    - Finally we will select the metadata to be the header and multi index, and put an index on the bacteria column. 
    - We will check after if the bacteria index is indeed unique. 

### Load Data files into a DataFrame


```python
# Write your answer here
import glob
import pandas as pd
path = DATA_FOLDER + 'microbiome/' # use your path
# verbose if want filenames, length of dataframe, and number of unique values
verbose = 0

allFiles = []
allFiles = glob.glob(path + "/MID*.xls")
if(verbose):
    print(allFiles)
# listDf = []
# pop last excel file from the list and build dataFrame 
firstFile = allFiles.pop()
totalDf = pd.read_excel(firstFile, index_col=None, header=None)

# get file name and remove extension to create column name MID...
fileName = firstFile.split('\\')[-1][:-4]

# Assign column names
totalDf.columns = ['bacteria', fileName]

testUnique = totalDf.bacteria
if(verbose):
    print(fileName)
    print(len(testUnique.value_counts()))
    print(len(totalDf))

totalDf.head()

# Import the rest of the files in the data frame and merge by doing an outer join
for file in allFiles:
#     Get file name to put as header
    fileName = file.split('\\')[-1][:-4]
    
    df = pd.read_excel(file,index_col=None, header=None)
    df.columns = ['bacteria', fileName]
    testUnique = df.bacteria
    if(verbose):
        print(fileName)
        print(len(testUnique.value_counts()))
        print(len(df))
#     listDf.append(df)
    totalDf = pd.merge(totalDf, df, how='outer')
    
# Fill NaN values with 'unknown' 
totalDf = totalDf.fillna('unknown')

# Order columns 
mid = ['bacteria'] + ['MID' + str(x) for x in range(1, 10)]
totalDf = totalDf[mid]
# Display DataFrame
totalDf
```

### Load metadata file

```python
metaDataFile = path + 'metadata.xls'
metaDataDf = pd.read_excel(metaDataFile)
```

```python
# metaDataDf[['GROUP', 'SAMPLE']]
metaDataDf
```

### Extract column as dictionnaries and create a DataFrame to match the columns of the total DataFrame

```python
group = metaDataDf['GROUP'].to_dict()
sample = metaDataDf['SAMPLE'].to_dict()
barcode = metaDataDf['BARCODE'].to_dict()

indexDict = {'BARCODE' : barcode.values(),'GROUP': group.values(), 'SAMPLE':sample.values()}

globDf = pd.DataFrame.from_dict(indexDict, 'index')
# Add bacteria column that will become our index in the end
globDf['bacteria'] = 'bacteria'

globDf
```

We assign here the row barcode as column names and remove this row. 

```python
globDf.columns = globDf.iloc[0]
globDf2 = globDf.reindex(globDf.index.drop('BARCODE'))
globDf2
```

We create the index for the columns 

```python
arrays = [list(metaDataDf['GROUP']),list(metaDataDf['SAMPLE']), list(metaDataDf['BARCODE'])]
# Add bacteria to the index 
arrays[0].insert(0,'bacteria')
arrays[1].insert(0, 'bacteria')
arrays[2].insert(0, 'bacteria')
# print(arrays)

# Concatenation metadata and data 
resultDf = pd.concat([globDf2, totalDf])
# Reordering of columns
mid = ['bacteria'] + ['MID' + str(x) for x in range(1, 10)]
resultDf = resultDf[mid]
resultDf.head()
```

```python
# Change column names and add the multi index on columns 
resultDf.columns = arrays
resultDf.columns.names= ['Group', 'Sample', 'Barcode']

# Set the bacteria as index for the rows and drop the bacteria column
resultDf_indexed = resultDf.set_index(['bacteria'])
resultDf_indexed.index.names = ['bacteria']
# Drop the bacteria column, to only leave it as index
resultDf2 = resultDf_indexed.drop(resultDf_indexed.index[[0,1]], 0)
resultDf2.drop('bacteria',1, inplace=True)
# Display preview of dataFrame 
resultDf2.head()
```
