```python
import json 
import os
from pprint import pprint
```

```python
filename = "../tags_wrong/non_main_tags_per_main_tag2.json"
file = open(filename, "r")
test = json.load(file)
```

```python
result_dict = {}
result_list = []
for key, value in test.items():
    print(key, value)
    temp_dict = {'tag' : key,
                'associated' : value}
    result_list.append(temp_dict)
result_dict['features'] = result_list
```

```python
# print(result_dict)
with open('../tags/data_tags.json', 'w') as fp:
    json.dump(result_dict, fp)
```

# New tree structure with name and children 

```python
result_dict = {}
result_list = []
for key, value in test.items():
#     print(key, value)
    list_subtag_json= []
    for subtag in value: 
        list_subtag_json.append({'level':'subtag',
                                 'name':subtag})
           
    temp_dict = {'level' :'tag',
                 'name' : key,
                'children' : list_subtag_json}
    result_list.append(temp_dict)
result_dict = {'level' :'toplevel',
               'name' : 'tags',
              'children' : result_list}
# pprint(result_dict)
with open('data_tags_tree.json', 'w') as fp:
    json.dump(result_dict, fp)
```
