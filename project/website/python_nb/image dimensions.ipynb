```python
import cv2
import os 
from pprint import pprint
import math
```

```python
folder = "images_duckduckgo"
list_filenames = []
for dirs, subdirs, files in os.walk(folder):
    for file in files: 
        list_filenames.append(file)
pprint(list_filenames)
```

```python
dim_max = 200
for filename in list_filenames:
    print(filename)
    path = os.path.join(folder,filename)
    test = cv2.imread(path)
    print(test.shape)
    (row, col, channel) = test.shape
    if(row >= col):
        row_final = dim_max
        col_final = math.floor(dim_max/row*col)
    else:
        col_final = dim_max
        row_final = math.floor(dim_max/col*row)
    cv2.imshow("normal", test)
    test2 = cv2.resize(test, (col_final,row_final))
    print(test2.shape)
    cv2.imshow("resized", test2)
#     cv2.waitKey(0)
    path_save = os.path.join("im_resized", filename)
    cv2.imwrite(path_save, test2)
cv2.destroyAllWindows()
```
