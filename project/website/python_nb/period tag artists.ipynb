```python
import json 
import os
import seaborn as sns 
from pprint import pprint
```

# Open json file

```python
folder = "../top_artists/"
list_filenames = []
for dirs, subdirs, files in os.walk(folder):
    for file in files:
        if(('top_artists_period_tag_artist' not in file) and ('top_artists_tag_period_artist' not in file)):
            list_filenames.append(file)
pprint(list_filenames)
```

# Get periods

```python
prefix = "translated_top_artist_tags_"
suffix = ".json"
def getPeriod(x):
    result = x.replace(prefix, "")
    result = result.replace(suffix, "")
    result = " - ".join(result.split("_"))
    return result
list_period = [getPeriod(x) for x in list_filenames]
pprint(list_period)
```

# Create dataset_array

```python
dataset_list = []
dataset_dict = {}
for index in range(0, len(list_filenames)):
    combo_dict = {"name" : list_period[index]}
    dataset_list.append(combo_dict)
dataset_dict['data'] = dataset_list
pprint(dataset_dict)
with open("../top_artists/period_names.json", 'w') as fp:
    json.dump(dataset_dict, fp)
```

# TAG -> PERIOD -> ARTISTS

```python
global_dict = {}
global_period_list = []
for filename in list_filenames:
    path = os.path.join(folder, filename)
    file = open(path, "r")
    json_content = json.load(file)
    period = getPeriod(filename)
    key_list = []
    for key, value in json_content.items():
        list_artists = []
        for artist in value:
            artist_dict = {"level" : "artist",
                          "name" : artist,
                          "children" : None}
            list_artists.append(artist_dict)
        key_dict = {"level": "genre",
                   "name" : key,
                   "children" : list_artists}
        key_list.append(key_dict)
    result_dict={"level" : "period",
                "name" : period,
                "children": key_list}
    global_period_list.append(result_dict)
global_dict = {"level" : "top",
              "name" : "global",
              "children" : global_period_list}
pprint(global_dict)
with open('../top_artists/top_artists_period_tag_artist.json', 'w') as fp:
    json.dump(global_dict, fp)
```

Invert 

# PERIOD -> TAG -> ARTISTS

```python
# Create new dictionary adapted 

new_dict = {}
for period in global_dict["children"]:
#     print(period)
    name_period = period['name']
    tag_list = period['children']
    for tag in tag_list:
        artists = tag['children']
        tag_name = tag['name']
        if(tag_name not in new_dict):
            new_dict[tag_name] = []
            
        new_dict[tag_name].append({'name':name_period,
                              'children' : artists,
                              'level' : 'period'})
# pprint(new_dict)

# Convert new dictionary in json format 
global_dict_new = {}
global_list_new = []
for tag,value  in new_dict.items():
    new_tag_dict = {'name' : tag,
                   'level' : 'tag',
                   'children' : value}
    global_list_new.append(new_tag_dict)
global_dict_new = {'name' :'global',
                  'level': 'top',
                  'children': global_list_new}
with open('../top_artists/top_artists_tag_period_artist.json','w') as fp:
    json.dump(global_dict_new, fp)
```
